# Ngram

```python
Xs = [[(dt,t,p) for dt,t,p in zip(dT,T,P)] for dT,T,P in zip(dataset["dtseqs"], dataset["Tseqs"], dataset["pitchseqs"])]
print len(Xs)
flatX = [x for X in Xs for x in X]
alphabet = list(set(flatX))
print len(alphabet), len(flatX)
```

```python
from utils.ngram import train
order = 3

model = train(Xs, alphabet, order)
#print model
```

```python
from utils.ngram import generate
xhat = generate(model,alphabet,order,nletters=300)
xhat = [alphabet[x] for x in xhat]
dtseq = [x[0] for x in xhat]
Tseq = [x[1] for x in xhat]
pitchseq = [x[2] for x in xhat]
midi = writeMIDI(dtseq, Tseq, pitchseq, path='../data/', label='ngram',tag="order#"+str(order))
```
